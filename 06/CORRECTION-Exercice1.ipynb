question 3)

L'idée est d'explorer les maillons de suivant en suivant... jusqu'à None


### a) avec une boucle while

```
def ajoute_a_la_fin(lst,v):
    m = lst
    # on explore les maillons de suivant en suivant jusqu'à None
    while m.suivant is not None:
        m = m.suivant
    m.suivant = Maillon(v)
    
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def __str__(self):
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '-' + str(self.suivant)

lst = Maillon(10, Maillon(20))
```

```
ajoute_a_la_fin(lst, 30)
```

### b) avec une fonction récursive

```
def ajoute_a_la_fin(lst,v):
    # cas de base
    if lst.suivant is None :
        lst.suivant = Maillon(v)
    # cas récursif
    else : 
        ajoute_a_la_fin( lst.suivant , v)


    
class Maillon:
    def __init__(self, val, suiv=None):
        self.valeur = val
        self.suivant = suiv

    def __str__(self):
        if self.suivant is None:
            return str(self.valeur)
        return str(self.valeur) + '-' + str(self.suivant)

lst = Maillon(10, Maillon(20))
```

```
ajoute_a_la_fin(lst, 30)
```
